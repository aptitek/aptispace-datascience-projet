### 2. Chargement des données nettoyées

```python
df = pd.read_csv('../data/processed/cleaned_data_sample.csv')
df.head()
```

### 3. Statistiques Descriptives

**À COMPLÉTER PAR L'ÉTUDIANT :**
Générez les résumés statistiques globaux et par groupes/catégories de votre jeu de données.

```python
# TODO: Utiliser df.describe() et des agrégationsgroupby()
df.describe()
# par quartier 
df.groupby("Neighborhood")["SalePrice"].agg([
    "mean", "median", "min", "max", "count"
]).sort_values(by="mean", ascending=False).head(10)
# par qualité globale
df.groupby("OverallQual")["SalePrice"].mean()
```

### 4. Ingénierie de variables (Feature Engineering)

**À COMPLÉTER PAR L'ÉTUDIANT :**
Appliquez la fonction `feature_engineering` de `src.data_clean` pour extraire des indicateurs temporels de base, et ajoutez d'autres variables dérivées complexes adaptées à votre problématique.

```python
df_feat = dc.feature_engineering(df, None)  # pas de timestamp ici

df_feat = df.copy()

df_feat["HouseAge"] = df_feat["YrSold"] - df_feat["YearBuilt"]
df_feat["TotalArea"] = df_feat["GrLivArea"] + df_feat["TotalBsmtSF"]

df_feat.head()
```

### 5. Analyse des Corrélations

**À COMPLÉTER PAR L'ÉTUDIANT :**
Analysez la matrice des corrélations des caractéristiques numériques à l'aide de Pandas.

```python
# TODO: Calculer les corrélations linéaires (Pearson ou Spearman)
correlations = df_feat.corr(numeric_only=True)

correlations
#heatmap propre
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(correlations, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
#top corrélations avec prix
print(
    correlations["SalePrice"]
    .sort_values(ascending=False)
    .head(10)
)
```
