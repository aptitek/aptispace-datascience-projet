# 📢 Étape 7 : Data Storytelling & Communication

Synthèse des résultats et restitution interactive pour les décideurs.

```
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/processed/cleaned_data_sample.csv')
print(f'Dataset charge : {df.shape}')
```

```
fig = px.scatter(
    df,
    x='GrLivArea',
    y='SalePrice',
    color='OverallQual',
    size='TotalBsmtSF',
    hover_data=['YearBuilt', 'GarageCars', 'zone_type'],
    color_continuous_scale='Viridis',
    labels={
        'GrLivArea': 'Surface habitable (sqft)',
        'SalePrice': 'Prix de vente ($)',
        'OverallQual': 'Qualite globale',
        'TotalBsmtSF': 'Sous-sol (sqft)'
    },
    title='Marche immobilier - Prix vs Surface'
)
fig.update_layout(plot_bgcolor='white', paper_bgcolor='white')
fig.show()
```

```
fig2 = px.box(
    df,
    x='OverallQual',
    y='SalePrice',
    title='Distribution des prix par qualite de construction',
    labels={'OverallQual': 'Qualite globale', 'SalePrice': 'Prix de vente ($)'}
)
fig2.update_layout(plot_bgcolor='white', paper_bgcolor='white')
fig2.show()
```

## Insights cles

1. La qualite de construction (OverallQual) est le facteur n1 - 57% importance
2. La surface habitable (GrLivArea) est le facteur n2 - 19% importance
3. Les biens de luxe sont identifiables visuellement - CNN 67% precision sur classe Luxe
4. Prix median : 163 000$ - Prix moyen : 180 921$ - ecart du aux outliers
