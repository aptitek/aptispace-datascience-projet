```python
import pandas as pd
import kagglehub
import os

import matplotlib.pyplot as plt
import seaborn as sns

# ==============================
# LOAD DATASET
# ==============================

# téléchargement dataset
path = kagglehub.dataset_download(
    "house-prices-advanced-regression-techniques"
)

# chemin train.csv
train_path = os.path.join(path, "train.csv")

# chargement données
df = pd.read_csv(train_path)

# aperçu
df.head()

# ==============================
# DISTRIBUTION DES PRIX
# ==============================

plt.figure(figsize=(10,5))

# histogramme
plt.hist(df["SalePrice"], bins=50)

# labels
plt.title("Distribution des prix des maisons")
plt.xlabel("Prix")
plt.ylabel("Nombre de maisons")

plt.show()

# ==============================
# SURFACE VS PRIX
# ==============================

plt.figure(figsize=(8,5))

plt.scatter(
    df["GrLivArea"],
    df["SalePrice"]
)

plt.title("Surface habitable vs Prix")
plt.xlabel("Surface habitable")
plt.ylabel("Prix")

plt.show()

# ==============================
# HEATMAP CORRÉLATIONS
# ==============================

# corrélations numériques
corr = df.corr(numeric_only=True)

# figure
plt.figure(figsize=(14,10))

# heatmap
sns.heatmap(corr)

plt.title("Matrice de corrélation")

plt.show()

# ==============================
# VARIABLES LES PLUS CORRÉLÉES
# ==============================

top_corr = corr["SalePrice"].sort_values(
    ascending=False
)

print(top_corr.head(10))

# ==============================
# QUALITÉ VS PRIX
# ==============================

plt.figure(figsize=(10,6))

sns.boxplot(
    x=df["OverallQual"],
    y=df["SalePrice"]
)

plt.title("Qualité générale vs Prix")

plt.show()

# ==============================
# BUBBLE CHART
# ==============================

plt.figure(figsize=(10,6))

plt.scatter(
    df["GrLivArea"],
    df["SalePrice"],

    # taille bulles
    s=df["OverallQual"] * 20,

    # couleur
    c=df["YearBuilt"],

    alpha=0.6
)

plt.xlabel("Surface habitable")
plt.ylabel("Prix")
plt.title("Bubble Chart Immobilier")

plt.colorbar(label="Année construction")

plt.show()
```
